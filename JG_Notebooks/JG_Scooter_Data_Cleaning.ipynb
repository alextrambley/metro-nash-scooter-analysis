Creating unique month dataframes with custom function

```python
def month_cleaner(month):
    """This will import csv files, create datetime columns, filter data, remove unecessary columns, create a unique month df."""
    
    #read in csv:
    csv = pd.read_csv('Data/' + str(month)+ '_trip.csv')
            
    #create datetime columns
    csv['pubTimeStamp'] = pd.to_datetime(csv.pubTimeStamp)
    csv['Month'] = csv.pubTimeStamp.dt.month
    csv['Day'] = csv.pubTimeStamp.dt.day
    csv['Hour'] = csv.pubTimeStamp.dt.hour
    
    csv['Miles'] = csv.tripDistance / 5280.0
    
    #filter columns
    csv = csv.loc[(csv.tripDuration > 1) & (csv.tripDuration < 1440)]
    csv = csv.loc[(csv.tripDistance > 1)]
    #csv = csv.loc[(csv.Miles < 150)]
    
    #create geometry column
    csv['start_geo'] = csv.apply(lambda x: Point((float(x.startLatitude	), float(x.startLongitude))), axis=1)
    csv['end_geo'] = csv.apply(lambda x: Point((float(x.endLatitude	), float(x.endLongitude))), axis=1)

    
    #drop unecessary columns
    csv = csv.drop(['pubTimeStamp','tripDistance', 'tripRoute', 'create_dt','startDate',	'startTime',	'endDate',	'endTime','tripRecordNum'], axis=1)
    
    #rearrange columns
#     csv = csv[['companyName', 'sumdID', 'Miles', 'tripDuration', 'Month', 'Day', 'Hour', 'start_geo', 'end_geo']]
    
    #create unique monthly dataframes
    if str(month) == 'may':
        global may
        may = csv.loc[csv.Month == 5]
        
    
    if str(month) == 'june':
        global june
        june = csv.loc[csv.Month == 6]
        
       
    if str(month) == 'july':
        global july
        global august
        july = csv.loc[csv.Month == 7]
        august = csv.loc[csv.Month == 8]
```

```python
def unique_company(df):
    """Finds all unique company names and creates seperate dataframes."""
    
    #variable calls 
    global lyft_df
    global bird_df
    global gotcha_df
    global spin_df
    global bolt_df
    global lime_df
    global jump_df
    
    #saving new df to variables
    lyft_df = df.loc[df.companyName == "Lyft"]
    bird_df = df.loc[df.companyName == "Bird"]
    gotcha_df = df.loc[df.companyName == "Gotcha"]
    spin_df = df.loc[df.companyName == "SPIN"]
    bolt_df = df.loc[df.companyName == "Bolt Mobility"]
    lime_df = df.loc[df.companyName == "Lime"]
    jump_df = df.loc[df.companyName == "JUMP"]
```
